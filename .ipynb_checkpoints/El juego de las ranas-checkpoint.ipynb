# Descripcin del juego

El juego es realmente un acertijo. Supongamos que tenemos un tablero como el siguiente:




Un tablero con 7 huecos, en los que los 3 huecos de la izquierda contienen fichas verdes, mientras que los 3 huecos de la derecha contienen fichas marrones.

El objetivo del juego/acertijo, es conseguir que todas las fichas verdes ocupen las posiciones de las fichas marrones, y viceversa. Para alcanzar el objetivo las fichas verdes slo pueden mover hacia la derecha mientras que las fichas marrones slo pueden mover hacia la izquierda.

Los movimientos permitidos son los siguientes:

Se puede mover una ficha una nica casilla hasta el hueco, que por supuesto deber ser contiguo:
    
    



O se permite avanzar dos casillas con una ficha, saltando otra ficha del color contrario:

## Implementando el juego

### main

El juego consiste en una inicializacin, en la que se crear el tablero, un bucle en el que el usuario va moviendo fichas hasta que no se puede mover nada ms, y un final en el que se comprueba si el usuario ha ganado o ha perdido.

Esto es bien fcil de implementar si se consigue alcanzar un nivel suficiente de abstraccin:


```python
def main():
    # Creamos el tablero de juego
    tablero = init()
    objetivo = tablero[::-1]

    # Vamos pasando turnos hasta que no podamos continuar
    while puedes_mover_algo(tablero):
        # Pintamos el tablero
        show(tablero)
        # Pedimos por teclado una ficha para mover
        pos = int(input('Selecciona una ficha para mover (por su posicin): '))
        # Nos aseguramos de que esa ficha puede mover
        if not puede_moverse(tablero, pos):
            print('The token cannot move!')
            continue
        # Y la movemos
        tablero = mover(tablero, pos)

    # Comprobamos si ha ganado
    if tablero == objetivo:
        print('FELICIDADES!')
    else:
        print('FRACASO!')
```

Visto de esta forma, el juego parece bien simple, cierto?

Efectivamente, al ejecutar la celda no nos da ningn error a pesar de que an no hemos creado las siguientes funciones:

    init()
    puedes_mover_algo(list)
    show(list)
    puede_moverse(list, int)
    mover(list, int)

Eso se debe a que al definir la funcin sta no se ejecuta todava.

As pues debemos ir definiendo esas funciones. Podemos empezar por init(), que es la ms sencilla.

### init()

La creacin del tablero no es demasiado compleja. Tan slo hay que crear una lista con 7 nmeros enteros, en el que el 0 representa el hueco vaco, el 1 las fichas verdes, y el -1 las fichas marrones.


```python
def init():
    return [1] * 3 + [0] + [-1] * 3
```

Como se puede observar, una funcin tremendamente simple, hasta tal punto que no requiere de ninguna otra funcin para trabajar. Por tanto, podemos recuperar la lista de funciones pendientes de main(), y continuar por la ms sencilla, que probablemente es show()

### show(list)

Mostrar el tablero puede ser bastante complejo, pero en sta ocasin nos limitaremos simplemente a pedirle a Python que nos pinte la lista, sin preocuparnos por el formato:


```python
def show(tablero):
    print(tablero)
```

De las funciones que nos faltan, las dos pueden ser muy complejas si no nos esforzamos en abstraernos lo suficiente. Pero con un poquito de abstraccin ambas se vuelven sumamente simples. Empezemos por puedes_mover_algo(list)

### puedes_mover_algo(list)

En realidad, sta funcin slo debe recorrer la lista, y preguntar si alguien puede moverse. En caso de encontrar alguna ficha que pueda moverse, devolveremos una respuesta afirmativa:


```python
def puedes_mover_algo(tablero):
    for i in range(len(tablero)):
        if puede_moverse(tablero, i):
            return True
    return False
```

Conviene reparar en que sta funcin hace uso de otra funcin puede_moverse(list, int), que todava no hemos creado. Sin embargo, esa funcin ya la requeriamos en main(), as que realmente no aadimos ms trabajo, tn slo aprovechamos las ventajas de usar funciones.

Entre las funciones que an tenemos pendientes, puede_moverse(list, int) y mover(list, int), claramente mover(list, int) debe ser nuestra siguiente candidata.

### mover(list, pos)

En un principio la funcin mover(list, int) puede parecer ms compleja de lo que es, pero en realidad es una funcin muy sencilla si tenemos en cuenta que:

     Ya hemos comprobado que la ficha se puede mover
     Slo existe una nica posicin a la que mover, pues slo existe un hueco

As que en realidad esta funcin slo debe intercambiar los valores de la posicin elegida y del hueco.


```python
def mover(tablero, ficha):
    agujero = get_agujero(tablero)
    tablero[ficha], tablero[agujero] = tablero[agujero], tablero[ficha]
    return tablero
```

En sta ocasin, hemos aadido una nueva funcin que debemos implementar, get_agujero(list), pero que es bien sencilla.

### get_agujero(list)

sta funcin tan slo debe buscar donde se encuentra el 0 en la lista, que adems sabemos que ser nico.


```python
def get_agujero(tablero):
    for i,valor in enumerate(tablero):
        if not valor:
            return i
```

Finalmente, debemos enfrentarnos a la funcin ms complicada, puede_moverse(list, int).

### puede_moverse(list, pos)

sta funcin puede volverse muy complicada, as que conviene tomarse un tiempo para pensar una estrategia. En nuestro caso vamos a intentar aprovechar el hecho de que slo hay un hueco:


```python
def puede_moverse(tablero, ficha):
    # Excluimos por supuesto fichas fuera del tablero
    if not -1 < ficha < len(tablero):
        return False
    # Y excluimos tambien el hueco
    if not tablero[ficha]:
        return False
    # Vamos a ver donde esta el hueco
    agujero = get_agujero(tablero)
    # Si el hueco es contiguo, y esta en el lado correcto, entonces sabemos positivamente que podemos mover
    if ficha + tablero[ficha] == agujero:
        return True
    # Si no es el caso, la ficha esta obligada a saltar
    if (ficha + 2 * tablero[ficha] == agujero) and (tablero[ficha + tablero[ficha]] != tablero[ficha]):
        return True
    # Si la ficha no puede avanzar o saltar, entonces no se puede mover
    return False

```

Ya tenemos el juego listo, as que es momento de jugar!

```python
main()
```
