## Resolver ecuaciones matemáticas con SciPy

Ejemplo básico: métodos de Brent y de Newton

Aunque el método de la bisección es muy conocido porque conceptualmente es muy sencillo de entender, nos vamos a olvidar de él y vamos a utilizar en su lugar el método de Brent, que viene implementado en SciPy en la función scipy.optimize.brentq, ya que según la documentación es la mejor opción. El método de Brent es un algoritmo complicado que combina el método de la bisección, el de la secante e interpolación cuadrática inversa: como el método de la bisección, tiene convergencia garantizada siempre que demos un intervalo [a,b]
donde f(a)f(b)<0; es decir, que haya un cambio de signo en el intervalo.

En el caso en que conozcamos un valor próximo a la solución, seguiremos utilizando el método de Newton o el de la secante a través de la función scipy.optimize.newton, que selecciona uno u otro en el caso de que le pasemos la función derivada o no. Vamos a resolver este ejemplo:

$e^{x/3}cosx+10sin3x=x^{2}/4$

```python
x = np.linspace(0, 9, 100)
plt.plot(x, np.exp(x / 3) * np.cos(x) + 10 * np.sin(3 * x), x, x ** 2 / 4)
```

Antes que nada debemos definir la función que va a representar la ecuación. Todas las subrutinas de búsqueda de raíces en realidad buscan ceros de funciones, así que debemos escribir

$f(x)≡e^{x/3}cosx+10sin3x−x^{2}/4=0$

```python
def f(x):
    return np.exp(x / 3.0) * np.cos(x) + 10 * np.sin(3 * x) - x ** 2 / 4
```

Vamos a buscar primero la solución que vemos que está entre 4 y 5. Para ello, utilizamos el método de Brent:

```python
from scipy.optimize import brentq
sol1 = brentq(f, 4.0, 5.0)
print (sol1)  # 4.40989799172
```

Ahora si queremos buscar la solución que está cerca de 2, podemos probar el método de la secante:

```python
from scipy.optimize import newton
sol2 = newton(f, 2.0)
print (sol2)  # 2.17349784856
```

Más sencillo todavía estas son las dos soluciones:

```python
plt.plot(x, np.exp(x / 3) * np.cos(x) + 10 * np.sin(3 * x), x, x ** 2 / 4)
plt.plot(sol1,  np.exp(sol1 / 3) * np.cos(sol1) + 10 * np.sin(3 * sol1), sol1, sol1 ** 2 / 4,'ro',label='solución 1')
plt.plot(sol2,np.exp(sol2 / 3) * np.cos(sol2) + 10 * np.sin(3 * sol2), sol2, sol2 ** 2 / 4,'go',label='solución 2')
plt.legend()
```

## Integración númerica

Para integración numérica e integración de ecuaciones diferenciales, SciPy proporciona el paquete integrate.

```python
from scipy import integrate
```

### Ejemplo:

$f(x)=e^{-x^{2}}$

```python
def f(x):
    return np.exp(-x ** 2)
```

Y ahora simplemente utilizamos la función quad, que recibe como argumentos la función y los límites de integración.

```python
integrate.quad(f, 0, 5)
```

Se devuelve el resultado de la integración y una estimación del error cometido.

Puede darse el caso en que nuestra función dependa de un cierto número de parámetros:


$f(x)=Ae^{−Bx^{2}}$

En este caso, debemos seguir respetando que el primer argumento es la variable de integración, pero a partir de ahí podemos añadir todos los argumentos que queramos:

```python
def f(x, A, B):
    return A * np.exp(-B * x ** 2)
```

A la hora de integrar esta función, tendremos que usar el parámetro args para dar valores los argumentos extra de la función:


```python
integrate.quad(f, 0, 5, args=(1.0, 1.0))
```

### Integral doble

```python
from scipy.integrate import dblquad
area = dblquad(lambda x, y: x*y, 0, 0.5, lambda x: 0, lambda x: 1-2*x)
area
```

Para resolver ecuaciones de orden superior, habrá que realizar una reducción de orden:

$y′′+y=0$

$\mathbf{f}(\mathbf{y})=\pmatrix{y\\y'}'=\pmatrix{y'\\y''}=\pmatrix{y'\\-y}$

```python
def f(y, t):
    return np.array([y[1], -y[0]])
```

```python
t = np.linspace(0, 10)
```

```python
y0 = np.array([1.0, 0.0])
```

```python
sol = integrate.odeint(f, y0, t)
```

## Sistemas de ecuaciones no lineales

Si estamos en el caso de tener un sistema de ecuaciones no lineales, SciPy también tiene métodos para resolverlos, a través del paquete MINPACK. Utilizaremos la función root, que además soporta varios métodos de resolución en función de si proporcionamos el jacobiano del sistema o no.

Vamos a resolver este ejemplo:

Una vez más, tenemos que transformar este sistema en $\vec{f}(\vec{x}) = \vec{0}$, donde las flechas indican vectores. En este caso,

$\vec{x} = \begin{pmatrix} x \\ y \end{pmatrix}$

$\vec{f}(\vec{x}) \equiv \begin{pmatrix} x \cos{y} - 4 \\ x y - y - 5 \end{pmatrix} = \vec{0}$

El código en Python será:

```python
from scipy.optimize import root 
def f(x): 
    """Sistema de dos ecuaciones con dos incógnitas. """ 
    return [ x[0] * np.cos(x[1]) - 4, x[1]*x[0] - x[1] - 5 ] 
sol = root(f, [1, 1], jac=False)
sol
# Devuelve objecto Result print sol.x 
# Result.x contiene la solución 
```

## Optimización

Comenzamos con un ejemplo de minimización de función escalar simple (de una variable). Supongamos que queremos minimizar la siguiente función, que se traza entre x = -10 a x = 10. La función tiene el siguiente aspecto. Dentro del dominio de la función, tiene un mínimo global y un mínimo local.

función objetivo: 

### $\sin{x}e^{-0.1(x-0.6)^{2}}$

```python
plt.plot(x,scalar1(x))
plt.grid()
plt.xlabel(' Valores de x')
plt.ylabel('Valores de la función')
```

```python
def scalar1(x):
    return np.sin(x)*np.exp(-0.1*(x-0.6)**2)
```

```python
x=np.linspace(-10,10,100)
```

El código para determinar el mínimo global es extremadamente simple con SciPy. Podemos usar la función minimize_escalar en este caso.

```python
from scipy import optimize
result = optimize.minimize_scalar(scalar1)
result
```

```python
result['x']
```

```python
result = optimize.minimize_scalar(scalar1, bounds = (0,10),method='Bounded')
result['x']
```
